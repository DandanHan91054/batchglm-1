Generate some data:

```python
from api.models.glm_nb import Simulator

sim = Simulator(num_samples=2000, num_genes=100)
sim.generate_sample_description(num_batches=0)
sim.generate()
```

Run differential expression test:

```python
%%capture
import diffexpr as de

test = de.test(sim.data, full_formula="~ 1 + condition", reduced_formula="~ 1")
```

```python
pval = test.likelihood_ratio_test()
print(pval)
```
